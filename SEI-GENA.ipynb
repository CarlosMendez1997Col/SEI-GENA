# 1. Ministerio de Ciencia Tecnología e Innovación

```python
def data_minciencias():
    base_url = "https://minciencias.gov.co"
    url_objetivo = "https://minciencias.gov.co/convocatorias/todas"  
    try:
        resp = requests.get(url_objetivo, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        datos_minciencias = []
        filas = soup.find_all("tr")
        for fila in filas:
            cols = fila.find_all("td")
            if len(cols) >= 2:
                link_tag = cols[1].find("a")
                if link_tag and link_tag.has_attr('href'):
                    enlace = urljoin(base_url, link_tag['href'])
                    titulo = link_tag.get_text(strip=True)             
                    datos_minciencias.append({"Institucion": "MinCiencias", "Titulo": titulo, "Url": enlace})       
        df_minciencias = pd.DataFrame(datos_minciencias)
        return df_minciencias
    except Exception as e:
        print(f"Error {e}")
        return df_minciencias
minciencias = data_minciencias()
show(minciencias, search=True)
```

# 2. Ministerio de Ambiente y Desarrollo Sostenible

```python
def data_minambiente():
    url_objetivo = "https://regalias.minambiente.gov.co/invitacion-2025-2026/"
    datos_minambiente = [] 
    try:
        resp = requests.get(url_objetivo, timeout=15)
        resp.raise_for_status()
        soup = BeautifulSoup(resp.text, "html.parser")       
        vistos = set()
        bloques = soup.find_all("div", class_="wpb_wrapper")
        
        for bloque in bloques:
            h3 = bloque.find("h3")
            boton = bloque.find("a", class_="vc_btn3-style-modern") 
            if h3 and boton:
                titulo = h3.get_text(strip=True)
                enlace = boton.get('href')
                if enlace and len(titulo) > 10 and enlace not in vistos:
                    vistos.add(enlace)
                    datos_minambiente.append({"Institucion": "MinAmbiente","Titulo": titulo,"Url": enlace})
        
        if not datos_minambiente:
            for t in soup.find_all("h3"):
                link = t.find_next("a", href=True)
                if link and "regalias" in link['href']:
                    titulo = t.get_text(strip=True)
                    enlace = link['href']
                    datos_minambiente.append({"Institucion": "MinAmbiente","Titulo": titulo,"Url": enlace})  

    except Exception as e:
        print(f"Error: {e}")
    df_minambiente = pd.DataFrame(datos_minambiente)
    return df_minambiente

minambiente_df = data_minambiente()
show(minambiente_df, search=True) 
```

# 3. Fundación Natura Colombia

```python
def data_natura():
    url_objetivo = "https://natura.org.co/convocatorias/"
    base_url = "https://natura.org.co"  
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
        resp = requests.get(url_objetivo, headers=headers, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        datos_natura = []
        enlaces_vistos = set()
        for a in soup.find_all("a", href=True):
            href = a['href']
            if "convocatoria" in href.lower() and href not in enlaces_vistos:
                enlace = urljoin(base_url, href)
                slug = urlparse(enlace).path.strip('/')
                titulo = slug.replace('-', ' ').lower()
                titulo = titulo.replace('convocatoria', '').strip().capitalize()       
                datos_natura.append({"Institucion": "Fundacion Natura", "Titulo": titulo, "Url": enlace})
                enlaces_vistos.add(href)
            df_natura = pd.DataFrame(datos_natura)
        return df_natura
    except Exception as e:
        print(f"Error {e}")
        return df_natura
fundacion_natura = data_natura()
show(fundacion_natura, search=True)
```

# 4. Fundación para la Conservación y el Desarrollo

```python
def data_fcds():
    url_objetivo = "https://fcds.org.co/adquisiciones/"
    datos_fcds = []
    
    try:
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}
        resp = requests.get(url_objetivo, headers=headers, timeout=20)
        resp.raise_for_status()
        soup = BeautifulSoup(resp.text, "html.parser")
        enlaces_pdf = soup.find_all("a", href=True)
        
        for link in enlaces_pdf:
            enlace = link['href']
            if "/wp-content/uploads/" in enlace and enlace.endswith(".pdf"):
                titulo = link.get_text(strip=True)
                if not titulo or len(titulo) < 5:
                    padre_h3 = link.find_parent("h3")
                    if padre_h3:
                        titulo = padre_h3.get_text(strip=True)

                if len(titulo) > 10 and "MÁS INFORMACIÓN" not in titulo.upper():
                    if not any(d['url'] == enlace for d in datos_fcds):
                        datos_fcds.append({"Institucion": "FCDS","Titulo": titulo,"Url": enlace})

    except Exception as e:
        print(f"Error: {e}")
    
    return pd.DataFrame(datos_fcds)

fcds_df = data_fcds()
show(fcds_df, search=True)
```

# 5. FontAgro

```python
def data_fontagro():
    url_objetivo = "https://fontagro.org/es/iniciativas/convocatorias"
    base_url = "https://fontagro.org" 
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
        resp = requests.get(url_objetivo, headers=headers, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        datos_fontagro = []
        tarjetas = soup.find_all("h4", class_="initiative-title") 
        for titulo_tag in tarjetas:
            titulo = titulo_tag.get_text(strip=True)   
            link_tag = titulo_tag.find("a") if titulo_tag.find("a") else titulo_tag.find_next("a")        
            if link_tag and link_tag.has_attr('href'):
                href = link_tag['href']
                enlace = urljoin(base_url, href)
                datos_fontagro.append({"Institucion": "FontAgro","Titulo": titulo,"Url": enlace}) 
                df_fontagro = pd.DataFrame(datos_fontagro)
        if not datos_fontagro:
            print("No se detectaron convocatorias")
        return df_fontagro
    except Exception as e:
        print(f"Error {e}")
        return df_fontagro
fontagro = data_fontagro()
show(fontagro, search=True)
```

# 8. Fondo Acción

```python
def data_fondo_accion():
    url_objetivo = "https://fondoaccion.org/convocatorias/"    
    try:
        driver.get(url_objetivo)   
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CLASS_NAME, "fg-item-title")))
        soup = BeautifulSoup(driver.page_source, "html.parser")
        datos_fondo = []
        enlaces_vistos = set()
        tarjetas_abiertas = soup.find_all("div", class_="eael-cf-abiertas")
        
        for tarjeta in tarjetas_abiertas:
            h2 = tarjeta.find("h2", class_="fg-item-title")
            if not h2: continue
            titulo = h2.get_text(strip=True)
            link_tag = h2.find_parent("a")
            if link_tag and link_tag.has_attr('href'):
                enlace = link_tag['href']
                if enlace not in enlaces_vistos and not enlace.lower().endswith(('.jpg', '.jpeg', '.png')):            
                    datos_fondo.append({"Institucion": "Fondo Acción","Titulo": titulo,"Url": enlace})
                    enlaces_vistos.add(enlace)
                    df_fondo = pd.DataFrame(datos_fondo)
        return df_fondo
    except Exception as e:
        print(f"Error {e}")
        return df_fondo
fondo_accion = data_fondo_accion()
show(fondo_accion, search=True)
```

# 9. Patrimonio Natural

```python
def data_patrimonio_natural():
    url_objetivo = "https://patrimonionatural.org.co/convocatorias/"
    base_url = "https://patrimonionatural.org.co"   
    try:
        driver.get(url_objetivo)
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CLASS_NAME, "elementor-heading-title"))) 
        
        soup = BeautifulSoup(driver.page_source, "html.parser")
        datos_patrimonio = []
        enlaces_vistos = set()
        titulos = soup.find_all("div", class_="elementor-widget-heading")
        for bloque in titulos:
            link_tag = bloque.find("a")
            if link_tag and link_tag.has_attr('href'):
                enlace = urljoin(base_url, link_tag['href'])
                widget_desc = bloque.find_next_sibling("div", class_="elementor-widget-text-editor")
                if widget_desc:
                    titulo = widget_desc.get_text(strip=True)
                else:
                    titulo = link_tag.get_text(strip=True)
                if len(titulo) > 10 and not enlace.lower().endswith(('.jpg', '.png', '.jpeg')):
                    if enlace not in enlaces_vistos:                
                        datos_patrimonio.append({"Institucion": "Patrimonio Natural","Titulo": titulo,"Url": enlace})
                        enlaces_vistos.add(enlace)
                        df_patrimonio = pd.DataFrame(datos_patrimonio)
        return df_patrimonio
    except Exception as e:
        print(f"Error {e}")
        return df_patrimonio
patrimonio_natural = data_patrimonio_natural()
show(patrimonio_natural, search=True)
```

# 10. Banco de Desarrollo de América Latina (CAF)

```python
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

def data_caf():
    fuentes = {"Convocatorias": "https://www.caf.com/es/trabaja-con-nosotros/convocatorias/","Licitaciones": "https://www.caf.com/es/trabaja-con-nosotros/licitaciones/"}
    base_url = "https://www.caf.com"
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
    lista_final = []
    for nombre_seccion, url_base in fuentes.items():        
        try:
            resp_inicial = requests.get(url_base, headers=headers, verify=False, timeout=15)
            soup_inicial = BeautifulSoup(resp_inicial.text, "html.parser")
            paginas_html = soup_inicial.find_all("a", href=True, class_="no-underline")
            numeros = [int(a.text) for a in paginas_html if a.text.isdigit()]
            total_paginas = max(numeros) if numeros else 1
            for i in range(1, total_paginas + 1):
                url_pag = f"{url_base}?page={i}"
                resp = requests.get(url_pag, headers=headers, verify=False, timeout=15)
                soup = BeautifulSoup(resp.text, "html.parser")
                encabezados = soup.find_all("h3", class_=["card__title", "card--gris-50__title"])
                for h3 in encabezados:
                    link_tag = h3.find("a", href=True)
                    if link_tag:
                        titulo = link_tag.get_text(strip=True)
                        href = link_tag['href']
                        enlace = urljoin(base_url, href)
                        lista_final.append({"Institucion": "CAF","Titulo": titulo,"Url": enlace})
                time.sleep(0.2)
                
        except Exception as e:
            print(f" Error {e}")
    df_resultados = pd.DataFrame(lista_final)
    df_resultados = df_resultados.drop_duplicates(subset=['Url']).reset_index(drop=True)
    
    return df_resultados
df_caf = data_caf()
show(df_caf,search=True)
```

# 11. GIZ (Deutsche Gesellschaft für Internationale Zusammenarbeit)

```python
def data_giz():
    chrome_options = Options()
    driver = webdriver.Chrome(options=chrome_options)
    base_url = "https://ausschreibungen.giz.de"
    url_inicial = "https://ausschreibungen.giz.de/Satellite/company/welcome.do"
    lista_total = []
    try:
        driver.get(url_inicial)
        for pagina in range(1, 7):
            time.sleep(3)
            filas = driver.find_elements(By.XPATH, "//table[contains(@class, 'csx-new-table')]/tbody/tr")
            for fila in filas:
                try:
                    titulo = fila.find_element(By.CLASS_NAME, "word-break").text.strip()
                    enlace = fila.find_element(By.CLASS_NAME, "noTextDecorationLink").get_attribute("href")
                    lista_total.append({"Institucion": "GIZ","Titulo": titulo,"Url": enlace})
                except:
                    continue
            try:
                boton_next = driver.find_element(By.CSS_SELECTOR, "a.browseForward.waitClick")
                driver.execute_script("arguments[0].click();", boton_next)
                time.sleep(2)
            except:
                print("Se alcanzó la última página")
                break

    except Exception as e:
        print(f"Error: {e}")
    finally:
        driver.quit()

    return pd.DataFrame(lista_total)
df_giz = data_giz()
show(df_giz, search=True)
```

# 12. Programa de las Naciones Unidas para el Desarrollo (PNUD)

```python
def crear_driver():
    chrome_options = Options()
    chrome_options.add_argument("--start-maximized")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument("--no-sandbox")
    service = Service(ChromeDriverManager().install())
    return webdriver.Chrome(service=service, options=chrome_options)
def normalizar_texto(texto):
    if not texto:
        return ""
    texto_normalizado = unicodedata.normalize('NFD', texto)
    return "".join([c for c in texto_normalizado if unicodedata.category(c) != 'Mn'])

def aplanar_a_ascii_puro(input_file, output_file):
    print(f" Aplanando archivo y normalizando tildes: {input_file}")
    try:
        with open(input_file, 'r', encoding='utf-8') as f_in:
            contenido = f_in.read()
        contenido_normalizado = normalizar_texto(contenido)
        data_plana = contenido_normalizado.encode('ascii', 'ignore')        
        with open(output_file, 'wb') as f_out:
            f_out.write(data_plana)
        print(f" Archivo normalizado y convertido a ASCII: {output_file}")
    except Exception as e:
        print(f" Error en aplanado: {e}")

def ejecutar_proceso_gena_completo(df):
    archivo_sql = "universo_licitaciones.sql"
    archivo_ascii = "universo_licitaciones_ASCII.sql"
    tmp_dir_html = "temp_html_files"
    tmp_dir_pdf = "temp_pdf_files"
    
    for folder in [tmp_dir_html, tmp_dir_pdf]:
        if not os.path.exists(folder):
            os.makedirs(folder)

    with open(archivo_sql, "w", encoding='utf-8') as f_sql:
        f_sql.write("-- SEI-GENA: Script de carga masiva\n")
        f_sql.write('CREATE SCHEMA IF NOT EXISTS "GENA_Schema";\n')
        f_sql.write('CREATE TABLE IF NOT EXISTS "GENA_Schema".licitaciones (institucion TEXT, titulo TEXT, url TEXT, informacion TEXT);\n')
        f_sql.write('TRUNCATE TABLE "GENA_Schema".licitaciones;\n\n')

    print(f" Iniciando extracción masiva: {len(df)} registros.")
    driver = crear_driver()
    
    try:
        for idx, row in df.iterrows():
            if idx > 0 and idx % 100 == 0:
                print(f" Reiniciando navegador (Registro {idx})...")
                driver.quit()
                driver = crear_driver()

            intentos = 0
            exito = False
            url_actual = str(row['Url'])
            texto_limpio = ""
            
            while intentos < 2 and not exito:
                try:
                    if url_actual.lower().endswith(".pdf") or "pdf" in url_actual.lower() or "document.cfm" in url_actual.lower():
                        path_pdf = os.path.join(tmp_dir_pdf, f"convocatoria_{idx}.pdf")
                        try:
                            response = requests.get(url_actual, timeout=20, verify=False)
                            with open(path_pdf, "wb") as f_pdf:
                                f_pdf.write(response.content)
                            
                            with fitz.open(path_pdf) as doc:
                                text_chunks = [pagina.get_text() for pagina in doc]
                                texto_limpio = " ".join(text_chunks).strip()
                            
                            if not texto_limpio or len(texto_limpio) < 20:
                                texto_limpio = f"PDF detectado pero sin capa de texto legible. URL: {url_actual}"
                        except Exception as pdf_err:
                            texto_limpio = f"Error en PDF: {pdf_err}. URL: {url_actual}"
                    
                    else:
                        driver.get(url_actual)
                        time.sleep(6) 
                        html_raw = driver.page_source
                        soup = BeautifulSoup(html_raw, "html.parser")
                        for tag in soup(["script", "style", "nav", "footer", "header", "aside"]):
                            tag.decompose()
                        texto_limpio = soup.get_text(separator=' ', strip=True)
                        
                        with open(os.path.join(tmp_dir_html, f"registro_{idx}.html"), "w", encoding='utf-8') as f_h:
                            f_h.write(html_raw)

                    info_sql = texto_limpio.replace("'", "''")
                    inst_sql = str(row['Institucion']).replace("'", "''")
                    tit_sql = str(row['Titulo']).replace("'", "''")
                    url_sql = url_actual.replace("'", "''")

                    with open(archivo_sql, "a", encoding='utf-8') as f_sql:
                        query = f"INSERT INTO \"GENA_Schema\".licitaciones (institucion, titulo, url, informacion) VALUES ('{inst_sql}', '{tit_sql}', '{url_sql}', '{info_sql}');\n"
                        f_sql.write(query)
                        f_sql.flush()

                    print(f" [{idx+1}/{len(df)}] Procesado: {row['Institucion'][:15]}...")
                    exito = True

                except Exception as e:
                    print(f" Error registro {idx}: {e}")
                    intentos += 1
                    driver.quit()
                    driver = crear_driver()

    finally:
        driver.quit()
        for folder in [tmp_dir_html, tmp_dir_pdf]:
            if os.path.exists(folder):
                shutil.rmtree(folder)
        
        aplanar_a_ascii_puro(archivo_sql, archivo_ascii)
        print(f"\n PROCESO TERMINADO ")

ejecutar_proceso_gena_completo(df_unificado)
```

```python
print("Cargando motor semántico multilingüe...")
model_embed = SentenceTransformer('intfloat/multilingual-e5-small')

df['full_text'] = (
    "passage: " + df["institucion"].fillna("") + " - " + 
    df["titulo"].fillna("") + " - " + 
    df["informacion"].fillna("Sin descripción adicional")
)

print(f"Generando coordenadas para {len(df)} registros. Esto tardará unos segundos...")
embeddings = model_embed.encode(df['full_text'].tolist(), show_progress_bar=True)
print("Sistema indexado y listo para búsqueda.")
```

```python
def analizar_con_llama3(query_usuario, contexto_resultados):
    url = "http://localhost:11434/api/generate"
    
    prompt_final = f"""
    Eres el Analista Senior de SEI-GENA.
    Usuario busca: {query_usuario}
    Resultados encontrados:
    {contexto_resultados}
    
    INSTRUCCIÓN:
    1. Identifica el mejor 'match'.
    2. Justifica por qué es el mejor (en español).
    3. Si hay convocatorias en inglés, tradúcelas o resúmelas brevemente.
    Responde de forma ejecutiva y profesional.
    """
    
    payload = {
        "model": "llama3",
        "prompt": prompt_final,
        "stream": False,
        "system": "Responde siempre en español profesional."
    }
    
    try:
        response = requests.post(url, json=payload, timeout=40)
        return response.json().get('response', 'Error al generar síntesis.')
    except:
        return "⚠️ Error: ¿Está Ollama corriendo con Llama-3?"


salida_chat = widgets.Output()
entrada_texto = widgets.Text(placeholder='¿Qué buscas hoy? (Ej: Proyectos de agua en Latam)', layout={'width': '75%'})
boton_enviar = widgets.Button(description='Consultar IA', button_style='primary', icon='rocket')
historial = []

def ejecutar_consulta(b):
    query = entrada_texto.value.strip()
    if not query: return
    
    salida_chat.clear_output()
    with salida_chat:
        display(HTML("<p style='color: #0277BD;'>🧠 Buscando el mejor match...</p>"))
    

    query_vec = model_embed.encode([f"query: {query}"])
    sims = cosine_similarity(query_vec, embeddings)[0]
    top_idx = sims.argsort()[-3:][::-1]
    registros = df.iloc[top_idx]
    

    contexto_txt = ""
    for _, row in registros.iterrows():
        contexto_txt += f"- [{row['institucion']}]: {row['titulo']}\n"
    

    analisis_ia = analizar_con_llama3(query, contexto_txt)
    

    res_html = f"""
    <div style="background:#f9f9f9; border-left: 5px solid #2E7D32; padding:15px; border-radius:8px;">
        <b style="color:#1B5E20;">🎯 ANÁLISIS DE MATCH:</b><br>{analisis_ia}
    </div>
    <div style="margin-top:10px;">
        <b>📂 RESULTADOS:</b><br>
    """
    for _, row in registros.iterrows():
        res_html += f"• <b>{row['institucion']}</b>: <a href='{row['url']}' target='_blank'>{row['titulo']}</a><br>"
    res_html += "</div>"
    
    historial.append(("usuario", query))
    historial.append(("bot", res_html))
    entrada_texto.value = ""
    
    with salida_chat:
        salida_chat.clear_output()
        for rol, msg in historial:
            bg = "#E3F2FD" if rol == "usuario" else "#FFFFFF"
            align = "right" if rol == "usuario" else "left"
            display(HTML(f'<div style="background:{bg}; padding:12px; border-radius:10px; margin:8px; text-align:{align}; border:1px solid #ddd;">{msg}</div>'))

boton_enviar.on_click(ejecutar_consulta)
entrada_texto.on_submit(lambda x: ejecutar_consulta(None))

display(widgets.VBox([salida_chat, widgets.HBox([entrada_texto, boton_enviar])]))
```
